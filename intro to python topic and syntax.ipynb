```python
import random as r
num=r.randint(1,100)
guess=int(input("guess the number"))
lives=3
while guess!=num and lives>0:
    if guess>num:
        print("too big")
    else:
        print("too small")
    lives-=1
    print(f"you have {lives} live(s) left")
    guess=int(input("guess the number again"))
if guess==num:
    print(f"congrate {guess} is the number")
    print(f"total guesses equals {g_count}")
else:
    print("game over lives used up")
```

```python
class H2:
    species="homo sapiens"
    def __init__(self,nam,ag,coun):
        self.name=nam
        self.age=ag
        self.country=coun
    def __gt__(self,other):
        if self.age>other.age:
            return True
        else:
            return False
    def my_info(x):
        print(f"my name is {x.name}")
        print(f"i am {x.age} years old")
        print(F"i was born in {x.country}")
```

```python
class Student(H2):
    workplace="school"
    def __init__(self, name, age, country, SN):
        super().__init__(name, age, country)
        self.school_name=SN
    def my_info(x):
        print(f"my name is {x.name}")
        print(f"i am {x.age} years old")
        print(F"i was born in {x.country}")
        print(F"i am student of {x.school_name}")
```
