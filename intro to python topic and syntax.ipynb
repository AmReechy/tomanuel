```python
import random as r
num=r.randint(1,100)
guess=int(input("guess the number"))
g_count=1
while guess!=num:
    if guess>num:
        print("too big")
    else:
        print("too small")
    guess=int(input("guess the number again"))
    g_count+=1
print(f"congrate {guess} is the number")
print(f"total guesses equals {g_count}")
```

```python
import random as r
num=r.randint(1,100)
guess=int(input("guess the number"))
lives=3
while guess!=num and lives>0:
    if guess>num:
        print("too big")
    else:
        print("too small")
    lives-=1
    print(f"you have {lives} live(s) left")
    guess=int(input("guess the number again"))
if guess==num:
    print(f"congrate {guess} is the number")
    print(f"total guesses equals {g_count}")
else:
    print("game over lives used up")
```

```python
prime=[]
while len(prime)<=5:
    test_num=int(input("please enter a prime number"))
    half=test_num//2+1
    for n in range(2, half+1):
        if test_num%n==0:
            p(f"{test_num} is not a prime number")
            break
    else:
        p(f"{test_num} is a prime number")
        if test_num in prime:
            p(f"{test_num} is already used")
        else:
            prime.append(test_num)
p("the five prime numbers you enter are :")
p(prime)
```
